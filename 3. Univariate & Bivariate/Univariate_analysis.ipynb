```python
def qualQuan(dataset):
    quan = []
    qual = []
    for columnname in dataset.columns:
        if dataset[columnname].dtype == 'O':
            qual.append(columnname)
        else:
            quan.append(columnname)
    return quan, qual
```

```python
quan, qual = qualQuan(dataset)
```

```python
descriptive = pd.DataFrame(index = ["Mean", "Median", "Mode", "Q1:25%", "Q2:50%", "Q3:75%", "85%", "Q4:100%", "IQR", "1.5 rule", 
                                    "Lesser", "Greater", "Min", "Max", "Skewness", "Kurtosis"], columns = quan)
for columnname in quan:
    descriptive[columnname]["Mean"] = dataset[columnname].mean() 
    descriptive[columnname]["Median"] = dataset[columnname].median()
    descriptive[columnname]["Mode"] = dataset[columnname].mode()[0]
    descriptive[columnname]["Q1:25%"] = dataset.describe()[columnname]["25%"]
    descriptive[columnname]["Q2:50%"] = dataset.describe()[columnname]["50%"]
    descriptive[columnname]["Q3:75%"] = dataset.describe()[columnname]["75%"]
    descriptive[columnname]["85%"] = np.percentile(dataset[columnname],85)
    descriptive[columnname]["Q4:100%"] = dataset.describe()[columnname]["max"]
    descriptive[columnname]["IQR"] = descriptive[columnname]["Q3:75%"] - descriptive[columnname]["Q1:25%"]
    descriptive[columnname]["1.5 rule"] = 1.5 * descriptive[columnname]["IQR"]
    descriptive[columnname]["Lesser"] =  descriptive[columnname]["Q1:25%"] - descriptive[columnname]["1.5 rule"] 
    descriptive[columnname]["Greater"] = descriptive[columnname]["Q3:75%"] + descriptive[columnname]["1.5 rule"]
    descriptive[columnname]["Min"] = descriptive[columnname].min()
    descriptive[columnname]["Max"] = descriptive[columnname].max()
    descriptive[columnname]["Skewness"] = descriptive[columnname].skew()
    descriptive[columnname]["Kurtosis"] = descriptive[columnname].kurtosis()
```

```python
descriptive
```

#### Finding Outliers

```python
def FindOutliers(quan):
    lesser = []
    greater = []
    for columnname in quan:
        if descriptive[columnname]["Min"] < descriptive[columnname]["Lesser"]:
            lesser.append(columnname)
        if descriptive[columnname]["Max"] > descriptive[columnname]["Greater"]:
            greater.append(columnname)        
    return lesser, greater
```

```python
lesser, greater = FindOutliers(quan)
```

```python
print("Lesser:", lesser)
print("\nGreater:", greater)
```

#### Replace Outliers

```python
def replaceOutliers():
    for columnname in lesser:
        dataset[columnname][dataset[columnname] < descriptive[columnname]["Lesser"]] = descriptive[columnname]["Lesser"]
    for columnname in greater:
        dataset[columnname][dataset[columnname] > descriptive[columnname]["Greater"]] = descriptive[columnname]["Greater"]
    return lesser, greater
```

```python
lesser, greater = replaceOutliers()
```

#### 1. How many Employees are highly job satisfied?

```python
def freqtable(columnname, dataset):
    freqtable = pd.DataFrame(columns = ["Unique Values", "Frequency", "Relative Frequency", "Cumulative Frequency"])

    freqtable["Unique Values"] = dataset[columnname].value_counts().index
    freqtable["Frequency"] = dataset[columnname].value_counts().values
    freqtable["Relative Frequency"] = freqtable["Frequency"]/1470
    freqtable["Cumulative Frequency"] = dataset[columnname].cumsum()
    return freqtable
```

```python
freqtable("JobSatisfaction", dataset)
```

Out of 1470 Employees, 459 Employees are highly satisfied with their job. So they given the rating of 4.
569 Employees are not really satisfied with their job.

```python
from scipy.stats import ttest_ind, chi2_contingency

# Make a copy of dataset
df = dataset.copy()

# Identify target
target = "Attrition"   # (values: "Yes", "No")

# Split groups
Attrition_yes = df[df[target] == "Yes"]
Attrition_no = df[df[target] == "No"]

print("===== Numerical Features (t-test) =====")
for col in quan:
    t_stat, p_val = ttest_ind(Attrition_yes[col], Attrition_no[col])
    result = "Significant" if p_val < 0.05 else "Not Significant"
    print(f"{col}: p={p_val:.4f} → {result}")

print("\n===== Categorical Features (Chi-square) =====")
for col in qual:
    contingency = pd.crosstab(df[col], df[target])
    chi2, p, dof, expected = chi2_contingency(contingency)
    result = "Significant" if p < 0.05 else "Not Significant"
    print(f"{col}: p={p:.4f} → {result}")
```
