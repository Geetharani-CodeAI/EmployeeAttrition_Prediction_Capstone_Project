```python
def qualQuan(dataset):
    quan = []
    qual = []
    for columnname in dataset.columns:
        if dataset[columnname].dtype == 'O':
            qual.append(columnname)
        else:
            quan.append(columnname)
    return quan, qual
```

#### 1. How many Employees are highly job satisfied?

```python
def freqtable(columnname, dataset):
    freqtable = pd.DataFrame(columns = ["Unique Values", "Frequency", "Relative Frequency", "Cumulative Frequency"])

    freqtable["Unique Values"] = dataset[columnname].value_counts().index
    freqtable["Frequency"] = dataset[columnname].value_counts().values
    freqtable["Relative Frequency"] = freqtable["Frequency"]/1470
    freqtable["Cumulative Frequency"] = dataset[columnname].cumsum()
    return freqtable
```

```python
freqtable("JobSatisfaction", dataset)
```

Out of 1470 Employees, 459 Employees are highly satisfied with their job. So they given the rating of 4.
569 Employees are not really satisfied with their job.

#### How many employees had a overall good worklife balance?

#### What is the average salary of a Male Employer?
